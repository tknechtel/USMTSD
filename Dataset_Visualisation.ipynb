```python
#df = pd.read_csv("/content/drive/MyDrive/Masterarbeit/Thesis/boxplot_bee.csv",delimiter=";",decimal=",")
#df = pd.read_csv("/content/drive/MyDrive/Masterarbeit/Thesis/boxplot_cd1.csv",delimiter=";",decimal=",") #6
df = pd.read_csv("/content/drive/MyDrive/Masterarbeit/Thesis/boxplot_cd2.csv",delimiter=";",decimal=",") #6
#df = pd.read_csv("/content/drive/MyDrive/Masterarbeit/Thesis/boxplot_occ.csv",delimiter=";",decimal=",") #5
#df = pd.read_csv("/content/drive/MyDrive/Masterarbeit/Thesis/boxplot_syn1.csv",delimiter=";",decimal=",") #5
#df = pd.read_csv("/content/drive/MyDrive/Masterarbeit/Thesis/boxplot_syn2.csv",delimiter=";",decimal=",") #6

autoplait = df["Autoplait"]
autoplait=autoplait.dropna()
IGTS_DP = df["IGTS DP"]
IGTS_DP = IGTS_DP.dropna()
IGTS_TD = df["IGTS TD"]
IGTS_TD = IGTS_TD.dropna()
FLOSS1 = df["FLOSS 1"]
FLOSS1 = FLOSS1.dropna()
FLOSS2 = df["FLOSS 2"]
FLOSS2 = FLOSS2.dropna()
GGS = df["GGS"]
GGS = GGS.dropna()

columns = [autoplait, IGTS_DP, IGTS_TD, FLOSS1, FLOSS2, GGS]

fig = plt.figure(figsize =(12, 6))
 
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
ax.yaxis.grid(True, linestyle='-', which='major', color='grey',
               alpha=0.5)
ax.set(
    axisbelow=True,  # Hide the grid behind plot objects
    title='Dataset chickendance2: Boxplot grouped by Method and Regime Score',
    xlabel='Method',
    ylabel='Regime Score',
)



ax.boxplot(columns)
plt.xticks(fontsize=16)
box = ax.boxplot(columns, patch_artist=True, meanline=True, showmeans=True)
plt.xticks([1, 2, 3, 4, 5, 6], ["Autoplait","IGTS DP", "IGTS TD", "FLUSS1", "FLUSS2","GGS"], rotation=0)
#plt.show()
#plt.draw()

colors = []

for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

plt.savefig(data_img + 'new_boxplot_cd2.png',dpi=300)
```
